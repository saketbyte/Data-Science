```python
country_respondents = df['Country'].value_counts()
country_respondents
respondents_python=country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
respondents_python
```

```python
python_df = pd.concat([country_respondents, respondents_python],axis='columns',sort=False)
python_df
```

```python
python_df.rename(columns={'LanguageWorkedWith':'PythonResp','Country':'RespondentNum'},inplace=True)
python_df
```

```python
python_df['%age']=(python_df['PythonResp']/python_df['RespondentNum']) *100
```

```python
python_df.sort_values(by="%age",ascending=False,inplace =True)
python_df.head(55)
```
